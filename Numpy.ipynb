### Combinando Indexing [ , ] e slicing [ : : ]

```python
t
```

```python
t[:2, ::2] # linha zero, mas apenas os dois primeiros elementos
```

<div class="alert-warning" style="padding: 10px; border-radius: 4px; font-size: 120%">
 Fatiamentos (slices) so visualizaes do array original.  um novo array, mas no so novos dados, pois compartilham o mesmo endereo na memria. Sendo assim, modificar uma 'fatia' do array, afeta o seu original.
</div>

```python
t
```

```python
t[0,::] = 1
```

```python
t
```

### Como fazer uma cpia de um array?

#### Se fizermos:
#### b = a[ : : ] 

No  uma cpia! 'b' e 'a' compartilham o mesmo endereo

```python
b = a.copy()
b
```

```python
b[0] = 6
```

```python
b, a
```
